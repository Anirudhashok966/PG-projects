```python
import pandas as pd
```

# Upload/cleaning and imputation

```python
df=pd.read_csv('asset.csv');df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Index Price(2022)'] = pd.to_numeric(df['Index Price(2022)'], errors='coerce')
df['Index Price(2019)'] = pd.to_numeric(df['Index Price(2019)'], errors='coerce')
df['Index Price(2016)'] = pd.to_numeric(df['Index Price(2016)'], errors='coerce')
df['Index Price(2014)'] = pd.to_numeric(df['Index Price(2014)'], errors='coerce')
```

```python
df.info()
```

```python
df1= df.drop(['Index Name','S.No'],axis=1);df1.head()
```

```python
import numpy as np
from sklearn.impute import KNNImputer

def fill_nan_with_knn(df1):
    
    imputer = KNNImputer(n_neighbors=4)  #adjsut for neighbour val
    
   
    df_imputed = pd.DataFrame(imputer.fit_transform(df1), columns=df1.columns)
    
    return df_imputed

# fill nan
df_imputed = fill_nan_with_knn(df1)

#display with comparison
print("Original DataFrame:")
print(df1)
print("\nDataFrame after KNN imputation:")
print(df_imputed)
```

```python
df_imputed.head()
```

```python
a=df['Index Name'];a.head()
```

```python
df_n = pd.concat([a,df_imputed],axis=1);df_n.head()
```

```python
df_n.shape
```

```python
df_n.isnull().sum()
```

# Euclidian distance and Similarity matrix

```python
#using min-max scaler scale the data
from scipy.spatial.distance import pdist, squareform
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

dfs = pd.DataFrame(scaler.fit_transform(df_imputed), columns=df_imputed.columns);dfs.head()
```

```python
dfs.shape
```

```python
dist = pdist(dfs, metric='euclidean')

#simalirity kxk matrix
simmax = squareform(dist)

# Display the standardized DataFrame and similarity matrix
print("Standardized DataFrame:")
print(dfs)
print("\nSimilarity Matrix (based on Euclidean distance):")
print(simmax)
```

```python
simmax.shape
```

# Recommender

```python
df_n.head()
```

```python

df_n = df_n.rename(columns={'Index Name':'product'})
```

```python
df_n.head()
```

```python
#df = df.reset_index()
product_name = df_n['product']
product_name.shape
```

```python
indices = pd.Series(df_n.index, index=df_n['product'])
```

```python
#for euclidian

def recommend(product):
    idx = indices[product]
    sim_scores = list(enumerate(simmax[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=False)
    sim_scores = sim_scores[1:31]
    product_indices = [i[0] for i in sim_scores]
    return product_name.iloc[product_indices]
```

```python
df_n['product'].iloc[0]
```

```python
recommend('S&P BSE Sensex').head(10)
```
