```python
s_cl = float(input(('enter call long strike price:')))
s_cs = float(input(('enter call short strike price:')))
s_pl = float(input(('enter put long strike price:')))
s_ps = float(input(('enter put short strike price:')))

a = int(input('enter asset price:'))

c_l= float(input('enter call long price:'))
c_s= float(input('enter call short price:'))
p_l=float(input('enter put long price:'))
p_s=float(input('enter put short price:'))


q_cl=int(input('call long quantity:'))
q_cs=int(input('call short quantity:'))
q_pl=int(input('put long quantity:'))
q_ps=int(input('put short quantity:'))
```

```python
setup_cost = (c_l * q_cl - (c_s * q_cs) + p_l * q_pl - (p_s * q_ps));print('setup_cost of the strategy is:',setup_cost)
```

```python
st_range = range(int(s_cl-100),int(s_cl+100),5)

strike = []
for i in st_range:
    if i>0:
        strike.append(i)


ast_range = range(5000,7010,5)

asset =[]
for j in ast_range:
    if j>0:
        asset.append(j)
print(asset)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# call long option exercise:
payoff_cl=[]
cost_cl= c_l+ s_cl # exercise iff St>K such that St-k is a profit
for p in asset:
    if (s_cl)>p: 
        pay=0-c_l*q_cl
        print('No exercise of Call long option at',p, ' Our payoff is:',pay )
        payoff_cl.append(pay)
        

    else:
        pay=(p-c_l-s_cl)*q_cl #incl the option price_ST-K
        print('Call long option is exercised at',p,'Our payoff is:',pay)
        payoff_cl.append(pay)
```

```python
plt.figure(figsize=(10,10))
plt.plot(asset,payoff_cl,color='red')
plt.xlabel('asset price')
plt.ylabel('payoff')
plt.grid()
plt.plot( marker = 'o')
plt.legend(['Call Long Payoff'])

#we need better axis
```

```python
# call short option exercise:
payoff_cs=[]
for p in asset:
    if (s_cs)>p:
        pay=c_s*q_cs
        print('Call option is not exercised at',p, 'Our payoff is:',pay ) #the option buyer does not exercise- profit!
        payoff_cs.append(pay)
    else:
        pay=(-p+c_s+s_cs)*q_cs #-(St-k) is payoff
        print('Call option is exercise at',p,'Our payoff is:',pay)
        payoff_cs.append(pay)
```

```python
plt.figure(figsize=(10,10))
plt.plot(asset,payoff_cs,color='blue')
plt.xlabel('asset price')
plt.ylabel('payoff')
plt.grid()
plt.plot( marker = 'o')
plt.legend(['Call short Payoff'])
#we need better axis
```

```python
# put long option exercise:
payoff_pl=[]
cost_pl= p_l+ s_pl
for p in asset:
    if (s_pl)>p: #K>St
        pay=(s_pl-p_l-p)*q_pl  #exercised: payoff (K-st)
        print(' Put long option is exercised at',p, 'Our payoff is:',pay )
        payoff_pl.append(pay)
    else:
        pay=-p_l*q_pl #no exercised
        print('Put long option is not exercised at',p,'Our payoff is:',pay)
        payoff_pl.append(pay)
```

```python
plt.figure(figsize=(10,10))
plt.plot(asset,payoff_pl,color='green')
plt.xlabel('asset price')
plt.ylabel('payoff')
plt.grid()
plt.plot( marker = 'o')
plt.legend(['Put Long Payoff'])
#we need better axis
```

```python
# put short option exercise:
payoff_ps=[]
for p in asset:
    if (s_ps)>p: #(K>St) the put option buyer exercises the contract! Loss
        pay=(-s_ps+p_s+p)*q_ps #payoff: -(K-st)
        print('Put option is exercised at',p, 'Our payoff is:',pay )
        payoff_ps.append(pay)
    else:
        pay=p_s*q_ps
        print('Put option is not exercised at',p,'Our payoff is:',pay)
        payoff_ps.append(pay)
```

```python
plt.figure(figsize=(10,10))
plt.plot(asset,payoff_ps,color='orange')
plt.xlabel('asset price')
plt.ylabel('payoff')
plt.grid()
plt.plot( marker = 'o')
plt.legend(['Put Short Payoff'])
#we need better axis
```

```python
payoff_calls = []

for x,y in zip(payoff_cl,payoff_cs):
    payoff_calls.append(x+y)
print('payoff calls',payoff_calls)

```

```python
payoff_puts = []

for m,n in zip(payoff_pl,payoff_ps):
    payoff_puts.append(m+n)
print('payoff puts',payoff_puts)
```

```python
payoff_total = []

for ii,jj in zip(payoff_calls,payoff_puts):
    payoff_total.append(ii+jj)
print('payoff calls',payoff_total)
```
