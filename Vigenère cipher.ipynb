```python
alphabets = [chr(i) for i in range(65,91)];alphabets #using python character command
```

```python

alpp = np.array([list('ABCDEFGHIJKLMNOPQRSTUVWXYZ') for _ in range(26)]) #generate a 26x26 matrix with alphabets 


for row in range(alpp.shape[0]):  #shuffle it row wise
    np.random.shuffle(alpp[row])

alpp


```

# Rule

```python
#keystream is as long as the plaintext
#plaintext positioning is on the columns while the keystream positioning is on the rows
# example: pt = Anirudh and key is ASHOKAS, so the cipher would be; first col of plain text is letter A and the row from the
#key is A, the corresponding alphabet (wrt, col and row) in the table is simply the first letter of the ciphertext

```

# Encoding

```python
#encoding
plaintext = 'ANIRUDH'

KEY = 'ASHOKAS'

#find the cipher text.

ciphertext = ''
ct = []

for pchar, kchar in zip(plaintext, KEY):
    ciphertext = df.loc[kchar, pchar]
    print(ciphertext)
    
    ct.append(str(ciphertext))

print(ct)
```

# Decoding

```python
#decoding: simply use the row index of the key and then find the cipher text in the row. The matching column would be 
#the plain_text
ciphertextt = 'OIZBZCM'

supposed_plaintext = ''

KEY = 'ASHOKAS'

for cchar,kchar in zip(ciphertextt,KEY):
    for col in df.columns:
        if df.loc[kchar,col]==cchar:
            print(col)
            supposed_plaintext = supposed_plaintext+col
            break
            
print(supposed_plaintext)
```
