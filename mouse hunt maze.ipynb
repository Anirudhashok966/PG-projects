# Define moves, Path and All paths

```python
def is_valid_move(matrix, visited, row, col):
    num_rows, num_cols = matrix.shape  #length of rows and cols
    if row < 0 or row >= num_rows or col < 0 or col >= num_cols: #moves are not to be out of bound
        return False
    else:
        return matrix[row, col] != -100 and not visited[row, col]  #to ensure that the next state is not 
    #trap or already visited
```

```python
def find_paths(matrix, visited, row, col, path, paths):
    num_rows, num_cols = matrix.shape
    visited[row, col] = True #to mark a cell(i,j) that is already visited.
    path.append((row, col))#if state is visited, add the index to the path
    
    if matrix[row, col] == 100: #if end goal of 100 is reached
        paths.append(path.copy())
    else:
        moves = [(row-1, col), (row+1, col), (row, col-1), (row, col+1)] #try to move in all possible directions
        #directions: up, down, left, right.
        
        for move in moves:
            next_row, next_col = move #based on previous action
            if is_valid_move(matrix, visited, next_row, next_col):
                find_paths(matrix, visited, next_row, next_col, path, paths)
    
    visited[row, col] = False #mark current cell as unvisited and remove from the path (to  backtrack)
    path.pop() #.pop() results the last value in list.


```

```python
def find_all_paths(matrix, start_row, start_col):
    num_rows, num_cols = matrix.shape
    visited = np.zeros_like(matrix, dtype=bool) # a m*n array with zeros for to track the visited cells.
    
    paths = [] 
    find_paths(matrix, visited, start_row, start_col, [], paths)
    return paths

```
